```python
import pandas as pd
import os

folder_path = '/content'

all_files = [f for f in os.listdir(folder_path) if f.endswith('.csv') and 'uber' in f.lower()]

print(all_files)
```

```python
df_list = []

for file in all_files:
    file_path = os.path.join(folder_path, file)

    temp_df = pd.read_csv(file_path)
    temp_df['source'] = file
    df_list.append(temp_df)

df = pd.concat(df_list, ignore_index=True)

print(df.shape)
df.head()
```

```python
df.rename(columns={
    'Date/Time': 'datetime',
    'Pickup_date': 'datetime'
}, inplace=True)
```

```python
if 'Date/Time' in df.columns:
    df['datetime'] = df['Date/Time']

elif 'Pickup_date' in df.columns:
    df['datetime'] = df['Pickup_date']
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
print(list(df.columns))
```

```python
df = df[['datetime', 'Lat', 'Lon', 'Base']]
```

```python
import pandas as pd

df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
df = df.dropna(subset=['datetime'])
```

```python
df.head()
df.info()
```

```python
import matplotlib.pyplot as plt

daily_demand.plot(figsize=(10,5), title="Daily Uber Demand")
plt.show()
```

```python
daily_demand = df.set_index('datetime').resample('D').size()
daily_demand = daily_demand.to_frame(name='demand')

daily_demand.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(daily_demand.index, daily_demand['demand'])
plt.title("Daily Uber Demand Over Time")
plt.xlabel("Date")
plt.ylabel("Demand")
plt.show()
```

```python
daily_demand['day'] = daily_demand.index.day
daily_demand['month'] = daily_demand.index.month
daily_demand['weekday'] = daily_demand.index.weekday
```

```python
from sklearn.model_selection import train_test_split

X = daily_demand[['day', 'month', 'weekday']]
y = daily_demand['demand']

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, predictions)
print("MAE:", mae)
```

```python
plt.figure(figsize=(10,5))
plt.plot(y_test.values, label="Actual")
plt.plot(predictions, label="Predicted")
plt.legend()
plt.title("Actual vs Predicted Demand")
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(daily_demand['demand'], order=(5,1,0))
model_fit = model.fit()

forecast = model_fit.forecast(steps=7)
print(forecast)
```

```python
plt.figure(figsize=(10,5))
plt.plot(daily_demand['demand'], label='Actual')
plt.plot(forecast, label='Forecast', color='red')
plt.legend()
plt.title("Demand Forecast")
plt.show()
```
